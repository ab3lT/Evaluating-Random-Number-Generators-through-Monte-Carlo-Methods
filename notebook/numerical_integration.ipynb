```python
import numpy as np
from scipy.integrate import quad


f = lambda x: np.exp(-x**2)

# Monte Carlo Integration for 1D
def monte_carlo_1d(f, a, b, N):
    samples = np.random.uniform(a, b, N)
    integral_estimate = (b - a) * np.mean(f(samples))
    return integral_estimate

# True value of the integral using numerical integration
true_value, _ = quad(f, 0, 1)

# Vary N and compute estimates
N_values = [10**i for i in range(1, 7)]
estimates_1d = [monte_carlo_1d(f, 0, 1, N) for N in N_values]

# Multidimensional Monte Carlo Integration
f_2d = lambda x, y: np.exp(-(x**2 + y**2))

def monte_carlo_2d(f, a, b, N):
    x_samples = np.random.uniform(a, b, N)
    y_samples = np.random.uniform(a, b, N)
    integral_estimate = (b - a)**2 * np.mean(f(x_samples, y_samples))
    return integral_estimate

# Estimate 2D integral
true_value_2d = 0.5577462853253181  # Precomputed or reference value
estimates_2d = [monte_carlo_2d(f_2d, 0, 1, N) for N in N_values]
```

```python

import matplotlib.pyplot as plt

# 1D Convergence Plot
plt.figure(figsize=(10, 5))
plt.plot(N_values, estimates_1d, marker='o', label='Monte Carlo Estimate')
plt.axhline(y=true_value, color='r', linestyle='-', label='True Value')
plt.xscale('log')
plt.xlabel('N (log scale)')
plt.ylabel('Estimated Integral')
plt.title('1D Monte Carlo Integration Convergence')
plt.legend()
plt.grid()
plt.show()

# 2D Convergence Plot
plt.figure(figsize=(10, 5))
plt.plot(N_values, estimates_2d, marker='o', label='Monte Carlo Estimate (2D)')
plt.axhline(y=true_value_2d, color='r', linestyle='-', label='True Value (2D)')
plt.xscale('log')
plt.xlabel('N (log scale)')
plt.ylabel('Estimated Integral (2D)')
plt.title('2D Monte Carlo Integration Convergence')
plt.legend()
plt.grid()
plt.show()
```
